```python
# Extraindo o data set
df = pd.read_csv('dataBases/cat_breeds.csv')
df
```

```python
# separando as colunas
coluna = df['Breed;Age_in_years;Age_in_months;Gender;Neutered_or_spayed;Body_length;Weight;Fur_colour_dominant;Fur_pattern;Eye_colour;Allowed_outdoor;Preferred_food;Owner_play_time_minutes;Sleep_time_hours;Country;Latitude;Longitude']
df['Breed'] = coluna.str.split(';').str[0]
df['Age_in_years'] = coluna.str.split(';').str[1]
df['Age_in_months'] = coluna.str.split(';').str[2]
df['Gender'] = coluna.str.split(';').str[3]
df['Neutered_or_spayed'] = coluna.str.split(';').str[4]
df['Body_length'] = coluna.str.split(';').str[5]
df['Weight'] = coluna.str.split(';').str[6]
df['Fur_colour_dominant'] = coluna.str.split(';').str[7]
df['Fur_pattern'] = coluna.str.split(';').str[8]
df['Eye_colour'] = coluna.str.split(';').str[9]
df['Allowed_outdoor'] = coluna.str.split(';').str[10]
df['Preferred_food'] = coluna.str.split(';').str[11]
df['Owner_play_time_minutes'] = coluna.str.split(';').str[12]
df['Sleep_time_hours'] = coluna.str.split(';').str[13]
df['Country'] = coluna.str.split(';').str[14]
df['Latitude'] = coluna.str.split(';').str[15]
df['Longitude'] = coluna.str.split(';').str[15]
df = df.drop('Breed;Age_in_years;Age_in_months;Gender;Neutered_or_spayed;Body_length;Weight;Fur_colour_dominant;Fur_pattern;Eye_colour;Allowed_outdoor;Preferred_food;Owner_play_time_minutes;Sleep_time_hours;Country;Latitude;Longitude', axis=1)
```
